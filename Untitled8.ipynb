```python
connection.close() 
import pymysql
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine=create_engine("mysql+pymysql://{user}:{pw}@localhost/{db}"
                    .format(user="root",
                           pw="1234",
                           db='world_x'))
```

```python
df=pd.read_sql_table('city',con=engine)
```

```python
query='''select * from city'''
df=pd.read_sql_query(query,con=engine)
```

```python
query2="""select c.Code, c.Name, cl.Language, cl.IsOfficial
from country as c
left join countrylanguage as cl
on c.Code=cl.CountryCode
where c.Code='NGA';"""
```

```python
df1=pd.read_sql(query2,con=engine)
```

## How to write back to the database

```python
df1.to_sql("nigeria_lang",con=engine,index=False)
```
